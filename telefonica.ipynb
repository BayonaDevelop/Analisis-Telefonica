El redondeo del uso mensual de internet no está alineado con la especificación: estás redondeando la suma de MB a un MB entero. La facturación se basa en GB completos por mes (después de sumar los MB mensuales, convierte a GB y redondea hacia arriba). Esto cambiará los GB facturados y los ingresos.

[Junta los datos agregados en un DataFrame para que haya un registro que represente lo que consumió un usuario único en un mes determinado.]

```python
# Fusiona los datos de llamadas, minutos, mensajes e Internet con base en user_id y month

df_calls_tmp = df_calls_per_month.rename(columns={
    'call_year': 'year', 
    'call_month': 'month'
})

df_messages_tmp = df_messages_per_month.rename(columns={
    'message_year': 'year', 
    'message_month': 'month'
})

df_internet_tmp = df_internet_per_month.rename(columns={
    'session_year': 'year', 
    'session_month': 'month'
})

df_dates = pd.concat([
    df_calls_tmp[['year', 'month']],
    df_messages_tmp[['year', 'month']],
    df_internet_tmp[['year', 'month']]
])

df_dates = df_dates.drop_duplicates()
df_dates = df_dates.sort_values(by=['year', 'month']).reset_index(drop=True)

df_user_detail = df_users[['user_id', 'city', 'plan']].sort_values(by='user_id')
df_user_detail = df_user_detail.reset_index(drop=True)

df_invoice = df_user_detail.merge(df_dates, how='cross')
df_invoice = df_invoice.sort_values(by=['user_id', 'year', 'month']).reset_index(drop=True)

# Calls
df_invoice = df_invoice.merge(
    df_calls_tmp, 
    on=['user_id', 'year', 'month'], 
    how='left'
)
df_invoice[['total_calls', 'total_duration']] = df_invoice[['total_calls', 'total_duration']].fillna(0)
df_invoice['total_calls'] = df_invoice['total_calls'].astype(int)
df_invoice['total_duration'] = df_invoice['total_duration'].astype(int)

df_invoice = df_invoice.merge(
    df_messages_tmp[['user_id', 'year', 'month', 'total_messages']], 
    on=['user_id', 'year', 'month'], 
    how='left'
)
df_invoice['total_messages'] = df_invoice['total_messages'].fillna(0).astype(int)

df_invoice = df_invoice.merge(
    df_internet_tmp[['user_id', 'year', 'month', 'gb_used']], 
    on=['user_id', 'year', 'month'], 
    how='left'
).rename(columns={'gb_used': 'total_gb_used'})
df_invoice['total_gb_used'] = df_invoice['total_gb_used'].fillna(0).astype(int)
```

```python
# Añade la información de la tarifa

df_plans['gb_per_month_included'] = df_plans['mb_per_month_included'] / 1024
df_plans = df_plans.rename(columns={'plan_name': 'plan'})

mapping = df_plans.set_index('plan')[[
    'minutes_included', 
    'messages_included', 
    'gb_per_month_included', 
    'usd_per_minute', 
    'usd_per_message', 
    'usd_per_gb', 
    'usd_monthly_pay'
]].to_dict('index')
new_cols = df_invoice['plan'].map(mapping).apply(pd.Series)
df_invoice = pd.concat([df_invoice, new_cols], axis=1)
df_invoice = df_invoice.loc[:, ~df_invoice.columns.duplicated()]

print(df_invoice.head())
```

[Calcula los ingresos mensuales por usuario (resta el límite del paquete gratuito del número total de llamadas, mensajes de texto y datos; multiplica el resultado por el valor del plan de llamadas; añade la tarifa mensual en función del plan de llamadas). Nota: Dadas las condiciones del plan, ¡esto podría no ser tan trivial como un par de líneas! Así que no pasa nada si dedicas algo de tiempo a ello.]

```python
# Calcula el ingreso mensual para cada usuario

df_invoice['minutes_extra'] = np.where(
    df_invoice['total_duration'] > df_invoice['minutes_included'], 
    (df_invoice['total_duration'] - df_invoice['minutes_included']) * df_invoice['usd_per_minute'], 
    0
)

df_invoice['messages_extra'] = np.where(
    df_invoice['total_messages'] > df_invoice['messages_included'], 
    (df_invoice['total_messages'] - df_invoice['messages_included']) * df_invoice['usd_per_message'], 
    0
)

df_invoice['internet_extra'] = np.where(
    df_invoice['total_gb_used'] > df_invoice['gb_per_month_included'], 
    (df_invoice['total_gb_used'] - df_invoice['gb_per_month_included']) * df_invoice['usd_per_gb'], 
    0
)

df_invoice['Amount'] = df_invoice['minutes_extra'] + df_invoice['messages_extra'] + df_invoice['internet_extra'] + df_invoice['usd_monthly_pay']

print(df_invoice[['user_id', 'plan', 'year', 'month', 'Amount']].head())
```

[Prueba la hipótesis de que son diferentes los ingresos promedio procedentes de los usuarios de los planes de llamada Ultimate y Surf.]

```python
# Prueba las hipótesis

income_surf = df_invoice[df_invoice['plan'] == 'surf']['Amount']
income_ultimate = df_invoice[df_invoice['plan'] == 'ultimate']['Amount']

# Realizar la prueba t para dos muestras independientes (Prueba de Welch)
results = st.ttest_ind(income_surf, income_ultimate, equal_var=False)

alpha = 0.05

print(f"Media de ingresos Surf: ${income_surf.mean():.2f}")
print(f"Media de ingresos Ultimate: ${income_ultimate.mean():.2f}")
print(f'Valor p: {results.pvalue}')

if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula: Existe una diferencia significativa entre los ingresos de ambos planes.")
else:
    print("No podemos rechazar la hipótesis nula: Los ingresos promedio son estadísticamente similares.")
```
